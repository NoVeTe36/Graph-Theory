### Cliques

```python
import networkx as nx
import numpy as np
from lib.hamilton import *
from lib.basic_graph_matrix import *
import matplotlib.pyplot as plt
import time
```

```python
def find_cliques_n(graph, n):
    start_time = time.time()
    A = nx.find_cliques(graph)
    cliques = []
    for clique in A:
        if len(clique) == n:
            cliques.append(clique)
    end_time = time.time()
    time_taken = end_time - start_time
    return cliques, time_taken

d = 8
n = 30
G = nx.random_graphs.random_regular_graph(d, n, seed=20)

# cliques, time_taken = find_cliques_n(G, 3)

for i in range(3, 8):
    cliques, time_taken = find_cliques_n(G, i)
    print(f"Number of {i}-cliques: {len(cliques)}")

# print(cliques)

graph = nx.Graph()
graph.add_edges_from(G.edges)
nx.draw(graph, with_labels=True)
plt.show()
```

```python
d = 100
n = 1000
G = nx.random_graphs.random_regular_graph(d, n, seed=20)

time_complexity = []

for clique in range (1, 100):
    # calculate the time complexity
    time_complexity.append(find_cliques_n(G, clique)[1])

plt.plot(time_complexity)
```
